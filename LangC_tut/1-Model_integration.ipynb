OpenAI Integration


```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain.agents import create_agent

def get_weather(location):
    """Get the weather in a location."""
    return f"The weather in {location} is sunny."

agent = create_agent(
    model = "gpt-3.5-turbo",
    tools = [get_weather],
    system_prompt = "You are a helpful assistant.",
)

agent.invoke({"messages":[{"role": "user", "content": "whats the captial of india and its weather?"}]})
```

GROQ Integration

```python
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

1. Through init_chat_model

```python
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("How does the OCR work?")
print(response.content)
```

2. Through Groq Specific library

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("How does the OCR work?")
print(response.content)
```

Streaming the response instead of waiting for the complete response like in all the AI chatbots

```python
for i in model.stream("Whats A YOLO"):
    print(i.content , end="", flush=True)
```
