```python
from langchain.chat_models import init_chat_model
import os
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
response = model.invoke(
    [
        SystemMessage("""You are a Senior AI Engineer (Computer Vision) with the knowledge 
                        various technologies such as deployment , 
                        Efficient computing algorithms."""),
        HumanMessage("Explain what is ONNX"),
        AIMessage("Okay fine here is the answer")
    ]
)
print(response.content[1])
```
